# PART 2. TRANSFORMING THE DATA

```python
#Opening the csv file
tdata = pd.read_csv("final_temperaturas.csv", index_col = 0)
tdata
```

```python
#Mapping the columns for easier manipulation
tdata = tdata.rename(columns = {"0":"a", "1" : "b", "2" : "c", "3" : "d", "4" : "e", "5" : "f", "6" : "g", "7" : "h", "8" : "i"}) 
```

```python
#Creating the functions to extract the date
def clean_day(dat):
    if len(str(dat)) == 8:
        return int(str(dat)[0:2])
    elif len(str(dat)) == 7:
        return int(str(dat)[0])

def clean_month(dat):
    if len(str(dat)) == 8:
        return int(str(dat)[2:4])
    elif len(str(dat)) == 7:
        return int(str(dat)[1:3])
    
def clean_year(dat):
    if len(str(dat)) == 8:
        return int(str(dat)[4:])
    elif len(str(dat)) == 7:
        return int(str(dat)[3:])
```

```python
#Creating the date column

tdata["day"] = tdata["a"].apply(lambda x: clean_day(x))
tdata["month"] = tdata["a"].apply(lambda x: clean_month(x))
tdata["year"] = tdata["a"].apply(lambda x: clean_year(x))
tdata["date"] = pd.to_datetime(tdata[["day", "month", "year"]])
```

```python
#Creating the columns for maximum and minimum temperature
tdata["tmax"] = tdata["b"].apply(lambda x: float(x))
tdata["tmin"] = tdata["c"].apply(lambda x: float(x))
```

```python
#Creating the column for the city
tdata["f"] = tdata["f"].fillna("")
tdata["e"] = tdata["e"].fillna("")
tdata["g"] = tdata["g"].fillna("")
tdata["h"] = tdata["h"].fillna("")
tdata["i"] = tdata["i"].fillna("")
tdata["city"] = tdata["d"] + tdata["e"] + tdata["f"] + tdata["g"] + tdata["h"] + tdata["i"] 
```

```python
#Keeping the final columns
f_col = ["day", "month", "year", "tmax", "tmin", "city", "date"]
final_data2 = tdata[f_col]
final_data2
```

```python
#Now, there are some cities that do not have 365 observations. Thus, we neeed to remove those from the database. First, we identify those cities that have 365 obs.
gb_city = final_data2.groupby(final_data2["city"])
cit_365 = []
for i,y in gb_city:
    if y.size == 365*7:
        cit_365.append(i)
```

```python
#Keeping the cities with 365 observations
clean_data = final_data2[final_data2["city"].apply(lambda x: x in cit_365)]
```
